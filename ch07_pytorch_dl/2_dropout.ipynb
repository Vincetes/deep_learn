```python
import torch
from torch import nn
```

```python
x = torch.randint(1, 10, (10,), dtype=torch.float32)  #这是单元素的写法
```

```python
# 定义一个 Dropout 层，丢弃概率为 0.5
dropout = nn.Dropout(p=0.5)
```

```python
y = dropout(x)
```

```python
print("输入张量 x：", x)
print("经过 Dropout 层后的输出张量 y：", y)  #随机失活
```
