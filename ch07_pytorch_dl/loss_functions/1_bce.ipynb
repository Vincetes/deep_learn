```python
import torch
import torch.nn as nn
```

```python
#输入数据
input = torch.randn(3, 2) 
print(input)
pred = torch.sigmoid(input)  #经过sigmoid激活函数
print(pred)
```

```python
#目标值
target = torch.tensor([[0, 1], [1, 0], [0, 1]], dtype=torch.float)
print(target)
```

```python
#定义损失函数
loss = nn.BCELoss()
print(loss(pred, target))
```
