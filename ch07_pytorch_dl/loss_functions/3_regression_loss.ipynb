```python
import torch
import torch.nn as nn
```

```python
#定义数据
input = torch.randn(3, 5)  #假设有3个样本，5个类别
target = torch.randn(3, 5)  
print("输入张量 input：", input)
print("目标张量 target：", target)
```

```python
#MSE损失函数
loss = nn.MSELoss()
print("均方误差损失值：", loss(input, target))
```

```python
#MAE损失函数
loss = nn.L1Loss()
print("平均绝对误差损失值：", loss(input, target))
```

```python
#Smoothed L1损失函数
loss = nn.SmoothL1Loss()
print("平滑L1损失值：", loss(input, target))
```
