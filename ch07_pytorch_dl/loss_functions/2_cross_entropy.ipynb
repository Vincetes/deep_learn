```python
import torch
import torch.nn as nn
```

```python
# 1.目标值对顺序编码的标签
input = torch.randn(6, 8)  #假设有3个样本，5个类别
print("输入张量 input：", input)
pred = torch.softmax(input, dim=1)  #经过softmax激活函数
print(pred)
```

```python
#目标值，6个数据的类别标签（0-7）
target = torch.tensor([1, 0, 3, 7, 5, 2])
print("目标张量 target：", target)
```

```python
#定义损失函数
loss= nn.CrossEntropyLoss()
print("交叉熵损失值：", loss(pred, target))
print("交叉损失熵值（直接使用input）：", loss(input, target))  #CrossEntropyLoss内部集成了softmax函数
```

```python
# 2.目标值为一组概率(独热编码)
input = torch.randn(6, 8)
print("输入张量 input：", input)
```

```python
target = torch.randn(6, 8).softmax(dim=1)
print("目标张量 target：", target)
```

```python
loss = nn.CrossEntropyLoss()
print("交叉熵损失值（独热编码目标值）：", loss(input, target))
```
