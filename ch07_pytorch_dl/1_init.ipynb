```python
import torch
import torch.nn as nn
```

```python
#定义一个全连接层
linear = nn.Linear(5, 2)
```

```python
# 1.常数初始化
nn.init.zeros_(linear.weight)
print(linear.weight)
```

```python
#根据指定的值
nn.init.constant_(linear.weight, 0.5)
print(linear.weight)
```

```python
# 2.作eye初始化
nn.init.eye_(linear.weight)
print(linear.weight)
```

```python
# 3.正态分布初始化 
nn.init.normal_(linear.weight, mean=0, std=10)  #均值和方差
print(linear.weight)  
```

```python
#4.均匀分布初始化
nn.init.uniform_(linear.weight, a=0, b=1)  #区间
print(linear.weight)
```

```python
# 5. Xavier初始化
nn.init.xavier_normal_(linear.weight)
print(linear.weight)
```
