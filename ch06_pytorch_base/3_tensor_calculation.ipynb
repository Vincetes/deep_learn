```python
import torch
import numpy as np
```

```python
#1.四则运算  这个_，表示返回一个新的tensor
tensor1 = torch.randint(0, 10, (2, 3))
print(tensor1)
```

```python
print(tensor1 + 1) #不会修改原始数据
```

```python
print(tensor1.add(1)) #不会修改原始数据
```

```python
print(tensor1)
```

```python
#修改原始数据
tensor1.add_(10)
print(tensor1)
```

```python
print(tensor1.sub_(1))
```

```python
#幂函数
print(tensor1.pow_(2))
```

```python
tensor1 = tensor1.float()
#开方运算
print(tensor1.sqrt_())
```

```python
#计算张量的哈达玛积  对应位置相乘
tensor1 = torch.randint(0, 10, (2, 2))
print(tensor1)
tensor2 = torch.randint(0, 10, (2, 2))
print(tensor2)
print(tensor1 * tensor2)
```

```python
print(tensor1.mul(tensor2))
```

```python
#张量的矩阵乘法 matrix multiplication
tensor1 = torch.randint(0, 10, (2, 3))
print(tensor1)
tensor2 = torch.randint(0, 10, (3, 2))
print(tensor2)
print(tensor1.matmul(tensor2))
```

```python
#验证内存的共享和节省内存
X = torch.randint(0, 10, (2, 3))
print(X)
print(id(X))
```

```python
# X = X + 10  #X是指向了一个新的张量
X += 10
```

```python
print(X)
```

```python
print(id(X))
```

```python
X = torch.randint(0, 10, (3, 2, 4))
Y = torch.randint(0, 10, (3, 4, 1))
print(X.shape)
print(id(X))
X @= Y  #如果后续X不再使用，可以使用X[:] = X @ Y来减少内存开销
print(X.shape)
print(id(X))
```
