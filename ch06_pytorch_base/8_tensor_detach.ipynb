```python
import torch
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x.detach()
print(x)
print(y)
```

```python
print(id(x))
print(id(y))
```

```python
print(x.storage().data_ptr())
print(y.storage().data_ptr())    #其是共享内存中的同一份数据
```

```python
# print(x)
# y.zero_() #同时更改的
# print(x)
```

```python
#分别对x和y进行后续计算
z1 = x ** 2
z2 = y ** 2
print('z1:', z1)
print('z2:', z2)
```

```python
print(z1.requires_grad)
print(z2.requires_grad)
```

```python
z1.sum().backward()
```

```python
# z2.sum().backward()  #报错，因为y不需要梯度追踪，其计算图已经被切断
```

```python
x = torch.ones(2, 2, requires_grad=True)
y = x * x
print(x)
print(y)
```

```python
#分离y
u = y.detach()
print("U:", u)
```

```python
#让u参与张量的计算
z = u * x
print("Z:", z)
```

```python
#反向传播，计算梯度
z.sum().backward()  #这里只是汇总标量，并没有损失维度
h = torch.ones_like(x) 
print(h)
print(h.sum()) #这是一个标量
```

```python
print(x.grad)
```

```python
print(x.grad == u) #将u参数当成常数来计算了
```
