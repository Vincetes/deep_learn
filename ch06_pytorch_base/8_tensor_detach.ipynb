```python
import torch
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x.detach()
print(x)
print(y)
```

```python
print(id(x))
print(id(y))
```

```python
print(x.storage().data_ptr())
print(y.storage().data_ptr())    #其是共享内存中的同一份数据
```

```python
# print(x)
# y.zero_() #同时更改的
# print(x)
```

```python
#分别对x和y进行后续计算
z1 = x ** 2
z2 = y ** 2
print('z1:', z1)
print('z2:', z2)
```

```python
print(z1.requires_grad)
print(z2.requires_grad)
```

```python
z1.sum().backward()
```
