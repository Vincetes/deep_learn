```python
import torch
import numpy as np
```

6.4.1张量元素类型转换

```python
tensor1 = torch.tensor([1, 2, 3])
print(tensor1.dtype)
```

```python
#类型转换
tensor2 = tensor1.float()
print(tensor2.dtype)

#或者直接修改tensor1
tensor1 = tensor1.type(torch.float)
print(tensor1.dtype)
```

```python
tensor1 = tensor1.half()
print(tensor1.dtype)
```

6.4.2 Tensor和ndarray转换

```python
np.set_printoptions(precision=4)  #指定输出精度
torch.set_printoptions(precision=4)
#tensor -> ndarray
tensor1 = torch.rand(3, 4)
print(tensor1)
```

```python
ndarray = tensor1.numpy()
print(ndarray)
```

```python
tensor1[:, 0] = 10
print(tensor1)
print(ndarray)
```

```python
#使用copy方法避免共享内存
array2 = tensor1.numpy().copy()
print(tensor1)
tensor1[:, 0] = 5
print(tensor1)
print(array2)
```

```python
#2. ndarray -> tensor  还是一样使用copy可以避免共享内存
ndarray = np.random.rand(3, 4)
tensor1 = torch.from_numpy(ndarray)
print(tensor1)
```
