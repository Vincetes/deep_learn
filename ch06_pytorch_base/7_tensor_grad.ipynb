```python
import torch
```

```python
#定义数据
x = torch.tensor([10.0])
y = torch.tensor([3.0])
print(x)
print(y)
```

```python
#初始化参数
w = torch.rand(1, 1, requires_grad=True)
b = torch.rand(1, 1, requires_grad=True)
```

```python
#前向传播
z = w * x + b
print(z)
```

```python
#设置损失函数
loss = torch.nn.MSELoss()
loss_value = loss(z, y)
print(loss_value)  #打印损失值
print(loss_value.is_leaf)
```

```python
#反向传播
loss_value.backward()
```

```python
#查看梯度
print('w的梯度：', w.grad)
print('b的梯度：', b.grad)
```
