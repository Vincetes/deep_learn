```python
import torch   #本章节介绍相应函数
import numpy as np
```

```python
tensor1 = torch.randint(0, 10, (2, 5, 3))
print(tensor1)
```

```python
#求和
print(tensor1.sum())
```

```python
print(tensor1.sum(dim=1))  #针对第一个维度对应元素求和
```

```python
#mean求均值 必须是浮点数
print(tensor1.float().mean())
print(tensor1)
```

```python
#求解标准方差
print(tensor1.float().std())
```

```python
#求解最大值、最小值
print(tensor1.max(dim = 0))  #输出最大值 以及 对应索引号
print(tensor1.max(dim = 1))
print(tensor1.max(dim = 2))
print(tensor1.min())
```

```python
print(tensor1.argmin())  #argmin / argmax 返回的是**“沿指定方向最小/最大值的下标”**，而不是值本身。
print(tensor1.argmax())
```

```python
#去重
print(tensor1.unique())
```

```python
#排序 默认对于最后一个维度就是行进行排序
print(tensor1.sort())
```
