```python
import torch
```

```python
tensor1 = torch.randint(0, 10, (2, 3, 5))
print(tensor1)
```

6.8.1 交换维度

```python
#交换维度
print(tensor1.transpose(0, 1))
```

```python
print(tensor1.transpose(0, 2))  #交换维度理解似乎不够
```

```python
# 重新排列多个维度
print(tensor1)
print(tensor1.permute(0, 2, 1))
```

6.8.2 调整形状 reshape 和 view 

```python
tensor1 = torch.randint(1, 10, (3, 5, 4))
print(tensor1)
```

```python
print(tensor1.reshape(3, -1))
```

```python
print(tensor1.reshape(-1))
```

```python
#view
print(tensor1.is_contiguous()) #这玩意
```

```python
print(tensor1.view(-1, 10))
print(tensor1.shape)
print(tensor1)
```

```python
tensor1 = tensor1.permute(2, 1, 0)
print(tensor1)
```

```python
print(tensor1.is_contiguous())
```

```python
print(tensor1.contiguous().view(-1))
```

6.8.3 增加或者删除维度

```python
tensor1 = torch.tensor([1, 2, 3])
print(tensor1)
```

```python
#在0维上增加一个矩阵
print(tensor1.unsqueeze(dim=0))   
```

```python
print(tensor1.unsqueeze(dim=1))
```

```python
tensor1 = torch.randint(1, 10, (3, 4))
print(tensor1)
print(tensor1.unsqueeze(dim = 2))
```

```python
#减少维度
print(tensor1)
print(tensor1.squeeze(dim = 0))
tensor2 = torch.randint(1, 10, (1, 2, 3))
print(tensor2) #只有指定维度为1维的时候，才可以缩减维度
print(tensor2.squeeze(dim = 0))   
```

6.9 张量拼接

```python
#拼接 是对应x维度的大小变大了
tensor1 = torch.randint(1, 10, (2, 3, 4))
tensor2 = torch.randint(1, 10, (2, 3, 4))
print(tensor1)
print(tensor2)
```

```python
print(torch.cat((tensor1, tensor2), dim = 0)) #只改变拼接的维数  比如是dim=2，则最后dim1 + dim2
```

```python
#堆叠  是增加了一个总体维度，空间增加
tensor1 = torch.randint(1, 10, (2, 3, 4))
tensor2 = torch.randint(1, 10, (2, 3, 4))
print(tensor1)
print(tensor2)
```

```python
tensor3 = torch.stack([tensor1, tensor2], dim = 0)
print(tensor3)
print(tensor3.shape)
```
