```python
import torch
```

```python
#用两组数据对比
x1 = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
x2 = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
print("X1:", x1)
print("X2:", x2)
```

```python
y1 = x1.sigmoid()
y2 = x2.sigmoid()
print("Y1:", y1)
print("Y2:", y2)
```

```python
y1.sum().backward()
print("X1的梯度:", x1.grad)
```

```python
y2.sum().backward()
print("X2的梯度:", x2.grad)
```

```python
z1 = y1.data
z2 = y2.detach()
print("Z1:", z1)
print("Z2:", z2)
```

```python
print(z1.requires_grad)
print(z2.requires_grad)
```

```python
z1.zero_()
z2.zero_()
print(y1)
print(y2)
```
