```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (3, 5, 4))
print(tensor1)
```

```python
#1.简单索引
print(tensor1[0, 0, 1])
```

```python
print(tensor1[0, 1])
```

```python
print(tensor1[0])
```

```python
print(tensor1[1:])
```

```python
# 列表索引
print(tensor1[[0, 1], [0, 1]]) #似乎明白了
```

```python
print(tensor1[[0, 2], [0, 1], [2,1]])
```

```python
#布尔索引
#选取符合条件的行，首元素大于5
mask = tensor1[:, :, 0] > 5
print(mask)
```

```python
print(tensor1[mask])  #先取每行第一列的值 > 5 生成 True/False，再把 True 对应的整行抽出来
```

```python
#选取符合条件的列，第二个元素大于5
mask = tensor1[:, 1, :] > 5
print(mask)
```

```python
# print(tensor1[mask])
# 需要转置
tensor2 = tensor1.mT  #把张量最后两个维度互换
print(tensor2[ mask])
```

```python
#选取符合条件的矩阵
mask = tensor1[:, 1, 2] > 5
print(mask)
print(mask.shape)
```

```python
print(tensor1[mask])  #mask必须保证前面的维度一致
```
