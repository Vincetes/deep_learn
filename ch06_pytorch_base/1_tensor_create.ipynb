```python
import torch
import numpy as np
```

1.基本张量创建

(1)按内容创建张量 tensor

```python
tensor1 = torch.tensor(10)
print(tensor1)
print(tensor1.size())
print(tensor1.dtype)
```

```python
#传入一个一维数组
tensor2 = torch.tensor([1,2,3])
print(tensor2)
print(tensor2.size())
print(tensor2.dtype)
```

(2)按照维度创建张量 Tensor


```python
tensor1 = torch.Tensor(2, 3, 4)
print(tensor1)
print(tensor1.size())
print(tensor1.dtype)
```

```python
tensor2 = torch.Tensor(np.ndarray([2,3,4])) #nan也是属于随机值
print(tensor2)
print(tensor2.size())
print(tensor2.dtype)
```

(3)创建指定类型的张量

```python
# int类型
tensor1 = torch.IntTensor(2, 3)
tensor2 = torch.tensor([1,2,3], dtype=torch.int)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# long类型
tensor1 = torch.LongTensor(2, 3)
tensor2 = torch.tensor([1,2,3], dtype=torch.long)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# float类型
tensor1 = torch.FloatTensor(2, 3)
tensor2 = torch.tensor([1,2,3], dtype=torch.float)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
#bool类型
tensor1 = torch.BoolTensor(2, 3)
tensor2 = torch.tensor([1,2,3], dtype=torch.bool)
print(tensor1.dtype)
print(tensor2.dtype)
```

6.3.2 创建指定区间的张量

```python
tensor1 = torch.arange(0, 10, 2)
print(tensor1)
```

```python
tensor2 = torch.linspace(0, 10, 5)
print(tensor2)
```

6.3.3创建指定数值的张量

```python
tensor1 = torch.zeros(2, 3)
print(tensor1)
print(tensor1.dtype)
```

```python
tensor2 = torch.ones(2, 3)
print(tensor2)
print(tensor2.dtype)
```

```python
tensor3 = torch.full((2, 3), 10)
print(tensor3)
print(tensor3.dtype)
```

```python
tensor4 = torch.eye(3, 3) #创建对角线为1其余元素为0的指定大小的矩阵
print(tensor4)
print(tensor4.dtype)
```

6.3.4 创建随机数张量

```python
tensor1 = torch.rand(2, 3)
print(tensor1)
print(tensor1.dtype)
```

```python
#创建指定范围之内的
tensor2 = torch.randint(0, 10, (2, 3))
print(tensor2)
print(tensor2.dtype)
```

```python
#创建标准正态分布
tensor2 = torch.randn(2, 3)
print(tensor2)
```

```python
# 创建自定义分布的张量
tensor2 = torch.rand(2, 3, generator=torch.Generator().manual_seed(42))
#
print(tensor2)
```

```python
#随机重新排列
tensor1 = torch.randperm(10) # 生成打乱后的索引，再拿去重排别的张量
print(tensor1)
```

```python
#查看和指定随机数种子
print(torch.random.initial_seed())
```
